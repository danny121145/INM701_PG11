```python

import pandas as pd

# Drop the column named "id" and "
df.drop(["Unnamed: 32", "id"], axis=1, inplace=True)
df
```

```python

# Replace 'M' with 1 and 'B' with 0
df['diagnosis'] = df['diagnosis'].replace({'M': 1, 'B': 0})
df.head(10) # Displays the first 10 rows
```

```python
# Univariate Analysis
plt.hist(df['radius_mean'], bins=20)
plt.xlabel('Radius Mean')
plt.ylabel('Frequency')
plt.title('Histogram of Radius Mean')
plt.show()
```

```python
plt.hist(df['texture_mean'], bins=20)
plt.xlabel('Texture Mean')
plt.ylabel('Frequency')
plt.title('Histogram of Texture Mean')
plt.show()
```

```python
# Visualize the distribution of the target variable
# Visualize the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='diagnosis', data=df)
plt.title('Distribution of Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

Pairplot:
Creates a pairplot to visualize the relationships between selected features.
Each subplot shows the scatter plot of two features, color-coded by the diagnosis (benign or malignant).

```python
# Pairplot to visualize relationships between features
plt.figure(figsize=(12, 10))
sns.pairplot(df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'diagnosis']], hue='diagnosis')
plt.show()
```

```python
# Select two features for the scatter plot
feature1 = 'radius_mean'
feature2 = 'texture_mean'
# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=feature1, y=feature2, hue='diagnosis', data=df)
plt.title(f'Scatter Plot of {feature1} vs {feature2}')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='concavity_mean', y='compactness_mean', hue='diagnosis', data=df)
plt.title('Scatter Plot of Concavity Mean vs Compactness Mean')
plt.xlabel('Concavity Mean')
plt.ylabel('Compactness Mean')
plt.show()
```
